```python
df_out = pd.DataFrame(columns=['clas_hware_id', 'classroom_id'])
df2 = pd.DataFrame(columns=['clas_hware_id', 'classroom_id'])
```

```python
for colm in ['OSS 318', 'OSS 333', 'OSS 313',
       'OSS LL18', 'OSS LL13', 'OWS 275', 'OSS LL14', 'OSS 230', 'OSS 328',
       'OSS 120', 'OSS 121', 'OSS 122', 'OSS 123', 'OSS 124', 'OSS 127',
       'OSS 214', 'OSS 226', 'OSS 227', 'OSS 236', 'OSS 325', 'OSS 326',
       'OSS 327', 'OSS 329', 'OSS 428', 'OSS 431', 'OSS 432', 'OSS 434',
       'OSS 435', 'OSS LL06', 'OSS LL10', 'OSS LL15', 'OSS LL17', 'OSS LL19',
       'FDC 212', 'FDC 213', 'FDC 214', 'BIN LL02', 'BIN LL02A', 'FDC 307']:
    df2[['clas_hware_id', 'classroom_id']] = df[['Unnamed: 1', colm]]
    df_out = df_out.append(df2, ignore_index=True)
    
df_out.head(10)
```
